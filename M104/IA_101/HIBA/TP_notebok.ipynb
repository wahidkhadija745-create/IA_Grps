# Exercice 1: Analyse des Tendances Centrales

```python
df=pd.read_csv('/content/StudentsPerformance.csv')
```

```python
df
```

```python
liste = ["math score", "writing score", "reading score"]

print("Mean:")
for x in liste:
    print(f"{x} :", df[x].mean())

print("\nMedian:")
for x in liste:
    print(f"{x} :", df[x].median())

print("\nMode:")
for x in liste:
    print(f"{x} :", df[x].mode().values)
```

# Exercice 3: Quartiles et IQR

```python
scores = ["math score", "reading score", "writing score"]

print("Q1 (25%)")
print(df[scores].quantile(0.25))

print("\nMedian (50%)")
print(df[scores].quantile(0.50))

print("\nQ3 (75%)")
print(df[scores].quantile(0.75))
```

# Exercice 4: Visualisation des Données

```python
import plotly.express as px
```

```python
fig=px.histogram(df,x="race/ethnicity",y=["math score","reading score","writing score"])
fig.show()
```

```python
fig = px.box(df, y=["math score","reading score","writing score"])
fig.show()
```

```python
fig = px.bar(df, x='race/ethnicity', y=["writing score","reading score",'math score'])
fig.show()
```

```python
df_long = df.melt(
id_vars="race/ethnicity",
value_vars=["writing score", "reading score", "math score"],
var_name="Subject",
value_name="Score")
```

```python
fig = px.histogram(
df_long,
x="race/ethnicity",
y="Score",
color="Subject",
barmode="group",
marginal="rug")
fig.show()
```

```python
df_long = df.melt(
    id_vars="race/ethnicity",
    value_vars=["writing score", "reading score", "math score"],
    var_name="Subject",
    value_name="Score")
```

```python
df_avg = df_long.groupby(["race/ethnicity", "Subject"])["Score"].mean().reset_index()
fig = px.line(
    df_avg,
    x="race/ethnicity",
    y="Score",
    color="Subject",
    markers=True)
```

```python
fig.update_layout(
yaxis_title="Average Score",
xaxis_title="Race/Ethnicity")
fig.show()
```
