#### 1. Analyse Univariée

##### 1.1 Variables Quantitatives


```python
# a) Histogramme
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

data = np.random.normal(loc=50, scale=15, size=500)
plt.figure(figsize=(8,5))
sns.histplot(data, bins=20, kde=True, color='blue')
plt.title("Histogramme d'une variable quantitative")
plt.xlabel("Valeurs")
plt.ylabel("Fréquence")
plt.show()
```

L'histogramme permet de visualiser la distribution des valeurs d'une variable quantitative. La présence d'une courbe de densité permet d'estimer la forme de la distribution.


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Génération des données
np.random.seed(42)
salaire = np.random.normal(3000, 700, 300)  # echantion de taille 300 valeur aurtour de la Moyenne 3000, de dispersion écart-type 700
#np.random.normal est utilisé pour simuler des données ayant une répartition proche de la réalité (ex : salaires, notes, tailles, etc.).
# Tracé de l'histogramme
plt.figure(figsize=(8, 5))
sns.histplot(salaire, bins=20, kde=True, color='royalblue')
plt.xlabel("Salaire (€)")
plt.ylabel("Effectif")
plt.title("Répartition des salaires des employés")
plt.show()
```

```python
# b) Boxplot

plt.figure(figsize=(6,4))
sns.boxplot(x=data, color='orange')
plt.title("Boxplot d'une variable quantitative")
plt.show()
```

Le boxplot permet de détecter les valeurs aberrantes et d'observer la dispersion des données.


```python
notes = np.random.normal(12, 3, 100)  # Moyenne 12, écart-type 3


plt.figure(figsize=(5, 6))
sns.boxplot(y=notes, color='teal')
plt.ylabel("Notes des étudiants")
plt.title("Distribution des notes")
plt.show()
# Les notes varient entre 5 et 18, avec une médiane autour de 12. Quelques valeurs extrêmes en dessous de 5 indiquent des étudiants en difficulté.
```

```python
import plotly.graph_objects as go

# Définition des catégories (nœuds)
labels = [
    "IT", "RH", "Finance", "Marketing",  # Départements
    "Homme", "Femme",  # Genre
    "Promu", "Non promu",  # Promotion
    "Reste", "Part"  # Résultat final
]

# Définition des flux (relations entre catégories)
source = [0, 0, 1, 1, 2, 2, 3, 3,  # Départements → Genre
          4, 5, 4, 5,  # Genre → Promotion
          6, 6, 7, 7]  # Promotion → Résultat final

target = [4, 5, 4, 5, 4, 5, 4, 5,  # Départements → Genre
          6, 6, 7, 7,  # Genre → Promotion
          8, 9, 8, 9]  # Promotion → Résultat final

values = [30, 20, 15, 25, 18, 22, 12, 28,  # Effectifs par département et genre
          40, 50, 30, 20,  # Genre → Promotion
          45, 15, 25, 25]  # Promotion → Résultat final

# Création du diagramme Sankey
fig = go.Figure(go.Sankey(
    node=dict(
        label=labels,
        pad=15,  # Espacement entre les nœuds
        thickness=20,  # Épaisseur des nœuds
        color=["lightblue", "lightblue", "lightblue", "lightblue",  # Départements
               "orange", "pink",  # Genre
               "green", "red",  # Promotion
               "blue", "gray"]  # Résultat final
    ),
    link=dict(
        source=source,
        target=target,
        value=values
    )
))

fig.update_layout(title_text="Flux des employés dans l'entreprise", font_size=12)
fig.show()
```

Le département IT a plus d’hommes que de femmes, tandis que RH et Marketing ont plus de femmes.

Les femmes sont promues en plus grande proportion que les hommes.

```python
import numpy as np
import matplotlib.pyplot as plt

# Paramètres
n_clients = 10000
proba_accident = 0.05 #5% acc grave par ans
cost_accident = 10000 #le cout 

# Simulation : 1 si accident (coût 10 000€), 0 sinon
claims = np.random.choice([0, cost_accident], size=n_clients, p=[1-proba_accident, proba_accident])

# Calcul de la moyenne empirique cumulée
means = np.cumsum(claims) / np.arange(1, n_clients + 1)

plt.figure(figsize=(10, 5))
plt.plot(means, label="Coût moyen empirique")
plt.axhline(y=500, color='r', linestyle='--', label="Espérance théorique (500€)")
plt.xlabel("Nombre de clients")
plt.ylabel("Coût moyen (€)")
plt.title("Illustration du Théorème des Grands Nombres en Assurance")
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Paramètres de la simulation
n_tours = 10000  # Nombre de tours
mise = 1  # Mise par tour
gain = 35  # Gain si le numéro choisi est tiré (35 fois la mise)
probabilite_gagner = 1 / 37  # Probabilité de gagner (un seul numéro sur 37)
perte = -mise  # Perte (si on ne gagne pas)

# Simulation des gains et pertes
resultats = np.random.choice([gain, perte], size=n_tours, p=[probabilite_gagner, 1 - probabilite_gagner])

# Calcul de la moyenne empirique cumulée des gains
moyenne_empirique = np.cumsum(resultats) / np.arange(1, n_tours + 1)

# Tracé du graphique
plt.figure(figsize=(10, 5))
plt.plot(moyenne_empirique, label="Moyenne empirique des gains")
plt.axhline(y=0, color='r', linestyle='--', label="Espérance théorique (0)")
plt.xlabel("Nombre de tours")
plt.ylabel("Gains cumulés")
plt.title("Illustration du Casino Gagnant à Long Terme")
plt.legend()
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import pearsonr, linregress

X = np.array([100, 150, 200, 250, 300, 350, 400, 450, 500, 550])
Y = np.array([20, 30, 40, 50, 60, 65, 70, 80, 85, 95])
data= pd.DataFrame({"X":X,"Y":Y})
print(data.head())
corelation = data['X'].corr(data['Y'])
print('la valeur de la correlation est \n:',corelation)
X_mean = np.mean(X)
Y_mean = np.mean(Y)
# A Y= aX+b
a = np.sum((X - X_mean) * (Y - Y_mean)) / np.sum((X - X_mean) ** 2)
b = Y_mean - a * X_mean


print(f"Équation de la régression : Y = {a:.2f}X + {b:.2f}")

#B
slope, intercept, r_value, p_value, std_err = linregress(X, Y)
print(f"Équation de la régression : Y = {slope:.4f}X + {intercept:.4f}")

# X=150

Y_pred = a * 150 + b
print(f"Estimation pour X=150 : {Y_pred:.2f}")




plt.scatter(X, Y, label="Données réelles")
plt.plot(X, a*X + b, color='red', label="Régression linéaire")
plt.xlabel("Connexions")
plt.ylabel("Commandes")
plt.legend()
plt.show()
```

```python
import numpy as np
import plotly.graph_objects as go
from scipy.stats import linregress

# Données
X = np.array([100, 150, 200, 250, 300, 350, 400, 450, 500, 550])
Y = np.array([20, 30, 40, 50, 60, 65, 70, 80, 85, 95])

# Régression linéaire
slope, intercept, r_value, p_value, std_err = linregress(X, Y)

# Droite de régression
Y_pred = slope * X + intercept

# Création de la figure interactive avec Plotly
fig = go.Figure()

# Ajout des points de données
fig.add_trace(go.Scatter(x=X, y=Y, mode='markers', name='Données', marker=dict(size=8, color='blue')))

# Ajout de la ligne de régression
fig.add_trace(go.Scatter(x=X, y=Y_pred, mode='lines', name='Régression linéaire', line=dict(color='red')))

# Mise en forme du graphique
fig.update_layout(
    title="Régression Linéaire Interactive",
    xaxis_title="Connexions",
    yaxis_title="Commandes",
    hovermode="x unified"
)

# Affichage du graphique interactif
fig.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px

# Générer un dataset fictif
np.random.seed(42)
n = 1000  # Nombre de clients

# Âge suit une distribution normale centrée sur 35 ans avec un écart-type de 10
ages = np.random.normal(loc=35, scale=10, size=n).astype(int)
ages = np.clip(ages, 18, 65)  # Limiter l'âge entre 18 et 65 ans

# Catégorie socio-professionnelle (aléatoire)
categories = np.random.choice(["Étudiant", "Employé", "Cadre", "Retraité"], size=n, p=[0.2, 0.5, 0.2, 0.1])

# Nombre d'achats suit une loi de Poisson (moyenne de 5 achats par mois)
achats = np.random.poisson(lam=5, size=n)

# Créer le DataFrame
df = pd.DataFrame({"Âge": ages, "Catégorie": categories, "Achats": achats})

# Afficher les premières lignes
print(df.head())
```

```python
from scipy.stats import zscore
import numpy as np
import matplotlib.pyplot as plt

# Génération des transactions simulées
np.random.seed(42)
transactions = np.random.normal(500, 100, 1000)  # Transactions autour de 500€ avec sigma=100
transactions[990:] = np.random.uniform(2000, 5000, 10)  # Ajout de fraudes simulées
print("##############################", transactions.shape)
# Calcul des Z-scores
z_scores = zscore(transactions)
anomalies = np.where(np.abs(z_scores) > 3)  # Détection des transactions avec un Z-score > 3
print("\n ##########################", anomalies)
# Visualisation
plt.figure(figsize=(10, 5))
plt.hist(transactions, bins=50, alpha=0.7, label="Transactions normales")
plt.scatter(transactions[anomalies], np.zeros_like(transactions[anomalies]), color='red', label="Anomalies", marker='x')
plt.legend()
plt.xlabel("Montant des transactions (€)")
plt.ylabel("Fréquence")
plt.show()
```

Distribution Normale - Prédiction des prix immobiliers avec régression linéaire

- Contexte:

Les erreurs en régression linéaire sont souvent supposées suivre une distribution normale.

Si les erreurs ne sont pas normalement distribuées, cela peut affecter la qualité des intervalles de confiance et la prise de décision.

- Correction:

On utilise un dataset simulé de prix immobiliers.

On entraîne un modèle de régression linéaire et on vérifie si les erreurs suivent une loi normale.


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import norm, shapiro
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

# Génération des données simulées (prix en fonction de la surface)
np.random.seed(42)
surface = np.random.uniform(50, 200, 500)  # Surface en m²
prix = 3000 * surface + np.random.normal(0, 20000, 500)  # Prix en euros avec un bruit gaussien

# Création du DataFrame
df = pd.DataFrame({'surface': surface, 'prix': prix})

# Séparation des données
X_train, X_test, y_train, y_test = train_test_split(df[['surface']], df['prix'], test_size=0.2, random_state=42)

# Entraînement du modèle
model = LinearRegression()
model.fit(X_train, y_train)

# Prédictions et erreurs
y_pred = model.predict(X_test)
residuals = y_test - y_pred  # Erreurs

# Test de normalité des résidus (Shapiro-Wilk)
shapiro_test = shapiro(residuals)
print(f"p-value du test de normalité de Shapiro-Wilk : {shapiro_test.pvalue:.4f}")

# Visualisation de la distribution des erreurs
sns.histplot(residuals, kde=True, bins=20, stat="density", label="Résidus")
x_vals = np.linspace(min(residuals), max(residuals), 100)
plt.plot(x_vals, norm.pdf(x_vals, np.mean(residuals), np.std(residuals)), label="Normale théorique", color="red")
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

# Simulation : consommation énergétique de 30 maisons (kWh/m²/an)
np.random.seed(0)
n = 30
true_mean = 47  # moyenne réelle
true_std = 5    # écart-type réel
data = np.random.normal(loc=true_mean, scale=true_std, size=n)

# Moyenne et écart-type observés
sample_mean = np.mean(data)
sample_std = np.std(data, ddof=1)
stderr = sample_std / np.sqrt(n)  # Écart-type de la moyenne

# Intervalle de confiance à 95 % (loi de Student car σ inconnu)
t_value = stats.t.ppf(1 - 0.025, df=n-1)
ci_lower = sample_mean - t_value * stderr
ci_upper = sample_mean + t_value * stderr

# Affichage
plt.figure(figsize=(8, 5))
plt.hist(data, bins=8, color='skyblue', edgecolor='black', alpha=0.7)
plt.axvline(sample_mean, color='red', linestyle='--', label=f"Moyenne : {sample_mean:.2f}")
plt.axvspan(ci_lower, ci_upper, color='green', alpha=0.2, label=f"IC 95% : [{ci_lower:.2f}, {ci_upper:.2f}]")
plt.title("Consommation énergétique (kWh/m²/an) - Échantillon de 30 maisons")
plt.xlabel("Consommation")
plt.ylabel("Fréquence")
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Correction des questions statistiques en Python
import numpy as np
import pandas as pd
from scipy import stats
import matplotlib.pyplot as plt

# Donnees des series X et Y
X = np.array([1, 1, 3, 0, 0, 2, 3, 1, 3, 0, 5, 0, 0, 1, 1])
Y = np.array([4, 4, 2, 4, 2, 3, 2, 1, 5, 4, 3, 5, 5, 5, 0])

# 1. Moyenne et médiane
moyenne_X = np.mean(X)
mediane_X = np.median(X)
moyenne_Y = np.mean(Y)
mediane_Y = np.median(Y)

# 2. Variance, écart-type et étendue
variance_X = np.var(X, ddof=1)
ecart_type_X = np.std(X, ddof=1)
etendue_X = np.ptp(X)

variance_Y = np.var(Y, ddof=1)
ecart_type_Y = np.std(Y, ddof=1)
etendue_Y = np.ptp(Y)

# 3. Quartiles de X
Q1_X = np.percentile(X, 25)
Q2_X = np.percentile(X, 50)
Q3_X = np.percentile(X, 75)

# 4. Écart interquartile
EI_X = Q3_X - Q1_X

# 5. Coefficient de corrélation de Pearson
r = np.corrcoef(X, Y)[0, 1]

# 6-7. Régression linéaire
connexions = np.array([80, 100, 120, 150, 90, 110, 130, 140, 160, 170])
commandes = np.array([20, 25, 30, 40, 22, 28, 35, 38, 42, 45])

slope, intercept, r_value, p_value, std_err = stats.linregress(connexions, commandes)

# Equation de la droite : Y = slope * X + intercept
estim_150 = slope * 150 + intercept

# Affichage des résultats
print("--- Statistiques descriptives ---")
print(f"Moyenne X = {moyenne_X:.2f}, Médiane X = {mediane_X:.2f}")
print(f"Moyenne Y = {moyenne_Y:.2f}, Médiane Y = {mediane_Y:.2f}")
print(f"Variance X = {variance_X:.2f}, Écart-type X = {ecart_type_X:.2f}, Étendue X = {etendue_X}")
print(f"Variance Y = {variance_Y:.2f}, Écart-type Y = {ecart_type_Y:.2f}, Étendue Y = {etendue_Y}")
print(f"Q1 = {Q1_X}, Q2 = {Q2_X}, Q3 = {Q3_X}, Écart interquartile = {EI_X}")
print(f"Coefficient de corrélation r = {r:.2f}")

print("\n--- Régression linéaire ---")
print(f"Équation de la droite : Y = {slope:.2f}*X + {intercept:.2f}")
print(f"Prévision pour 150 connexions : {estim_150:.2f} commandes")
```
