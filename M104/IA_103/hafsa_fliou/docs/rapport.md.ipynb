**1. Explication des tendances observées**

L’analyse des données a permis d’identifier plusieurs tendances importantes.
Tout d’abord, une relation a été observée entre le nombre total de reviews et le score des jeux, suggérant que les jeux les plus populaires ont tendance à obtenir de meilleures évaluations.

De plus, certains genres de jeux présentent des scores plus élevés que d’autres, ce qui indique une préférence des utilisateurs pour certains types de jeux.

Enfin, les réductions (Discount_Pct) n’ont pas toujours un impact significatif sur le score, ce qui suggère que la qualité du jeu reste le facteur principal.

**2. Justification avec des preuves statistiques**

Les conclusions précédentes sont appuyées par plusieurs éléments statistiques.

La matrice de corrélation montre une relation très faible entre certaines variables quantitatives.
Les boxplots par genre mettent en évidence des différences de distribution des scores.
Les distributions (histogrammes) montrent que certaines variables suivent une tendance spécifique (par exemple une distribution asymétrique).

Ces analyses confirment que les relations observées ne sont pas aléatoires mais basées sur les données.

**3. Recommandations basées sur les données**

Sur la base des résultats obtenus, plusieurs recommandations peuvent être proposées :

Les développeurs devraient se concentrer sur les genres les mieux notés afin de maximiser la satisfaction des utilisateurs.
L’amélioration de la qualité globale des jeux semble plus importante que les stratégies de réduction de prix.
Les jeux compatibles avec Steam Deck pourraient être davantage valorisés si une corrélation positive avec les scores est observée.

Ces recommandations permettent d’orienter les décisions stratégiques dans l’industrie du jeu vidéo.

**4. Limites de l’analyse**

Malgré les résultats obtenus, certaines limites doivent être prises en compte :

La présence éventuelle de valeurs manquantes ou bruitées peut affecter la précision des résultats.
Les corrélations observées ne signifient pas nécessairement une relation de causalité.
Le dataset peut ne pas être représentatif de l’ensemble des jeux disponibles sur le marché.
L’incertitude statistique peut influencer certaines conclusions, notamment pour les catégories avec peu d’observations.

Il est donc important d’interpréter les résultats avec prudence.
