```python
import numpy as np
import pandas as pd
```

```python
series = {
    "Paire": [2, 4, 6, 8, 10, 12, 14, 16],
    "Impaire": [3, 5, 7, 9, 11, 13, 15, 17, 19],
    "Répétées": [1, 2, 2, 3, 3, 3, 4, 4, 5, 6, 6, 7]
}
```

```python
# Fonction pour calculer les quartiles manuellement
def manual_quartiles(data):
    n = len(data)
    Q1 = (data[(n // 4) - 1] + data[n // 4]) / 2 if n % 2 == 0 else data[(n // 4)]
    Q2 = (data[(n // 2) - 1] + data[n // 2]) / 2 if n % 2 == 0 else data[n // 2] #la médiane est la moyenne des deux valeurs autour de l'indice
    Q3 = (data[(3 * n // 4) - 1] + data[3 * n // 4]) / 2 if n % 2 == 0 else data[(3 * n // 4)]
    return Q1, Q2, Q3

```

```python
# Calcul des quartiles pour chaque methode
results = []
for name, data in series.items():# utilisée pour itérer à travers un dictionnaire series. Elle permet de parcourir chaque (Key and Value) du dictionnaire.
    # Méthode manuelle
    Q1_manual, Q2_manual, Q3_manual = manual_quartiles(data)
    
    # Méthode Pandas 
    data_series = pd.Series(data) # Cree un objet Series a partir de data similaire à un tableau ou une liste avec un index pour chaque valeur
    Q1_pandas = data_series.quantile(0.25)
    Q2_pandas = data_series.quantile(0.5)
    Q3_pandas = data_series.quantile(0.75)
    
    # Méthode NumPy
    Q1_numpy = np.percentile(data, 25)
    Q2_numpy = np.percentile(data, 50)
    Q3_numpy = np.percentile(data, 75)
    
    results.append([name, "Manuelle", Q1_manual, Q2_manual, Q3_manual])
    results.append([name, "Pandas", Q1_pandas, Q2_pandas, Q3_pandas])
    results.append([name, "NumPy", Q1_numpy, Q2_numpy, Q3_numpy])

```

```python
# Affichage sous forme de tableau
quartiles_df = pd.DataFrame(results, columns=["Série", "Méthode", "Q1", "Q2 (Médiane)", "Q3"])
quartiles_df
```

```python
# appliquer une style pour la table des resultats

quartiles_df = pd.DataFrame(results, columns=["Série", "Méthode", "Q1", "Q2 (Médiane)", "Q3"])

def style_table(df):
    return df.style.applymap(lambda x: 'background-color: lightgreen' if isinstance(x, (int, float)) else '', subset=["Q1", "Q2 (Médiane)", "Q3"]) \
                   .set_table_styles([{'selector': 'thead th', 'props': [('background-color', 'lightblue'), ('color', 'black')]},
                                      {'selector': 'tbody tr:nth-child(even)', 'props': [('background-color', 'lightgray')]},
                                      {'selector': 'tbody tr:nth-child(odd)', 'props': [('background-color', 'white')]},
                                      {'selector': 'tbody tr:hover', 'props': [('background-color', 'yellow')]}])

# Appliquer le style
quartiles_df_styled = style_table(quartiles_df)

# Afficher le DataFrame stylisé
quartiles_df_styled
```
