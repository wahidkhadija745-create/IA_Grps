```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("country_impact.csv")

print("Aperçu des données :")
print(df.head())


```

```python
print("\n--- STRUCTURE ---")
print(df.info())

print("\n--- VALEURS MANQUANTES ---")
print(df.isnull().sum())

print("\n--- STATISTIQUES (ANOMALIES) ---")
print(df.describe())

print("\n--- FRÉQUENCES ---")
print(df['Inflation_Risk'].value_counts())
print(df['Vulnerability'].value_counts())


```

```python
print("\n--- STATISTIQUES DESCRIPTIVES ---")
print(df.describe())

print("\n--- MOYENNE / MÉDIANE / ÉCART-TYPE ---")
print("Moyenne GDP Impact:", df['GDP_Impact_Pct'].mean())
print("Médiane GDP Impact:", df['GDP_Impact_Pct'].median())
print("Écart-type GDP Impact:", df['GDP_Impact_Pct'].std())
```

```python
df['GDP_Impact_Pct'].hist()
plt.title("Distribution GDP Impact")
plt.show()
```

```python
df.boxplot(column='GDP_Impact_Pct')
plt.title("Boxplot GDP Impact")
plt.show()
```

```python
df['Inflation_Risk'].value_counts().plot(kind='pie')
plt.title("Inflation Risk")
plt.show()
```

```python
print("\n--- CORRÉLATIONS ---")
print(df.corr(numeric_only=True))

print("\n--- RELATION Inflation / GDP ---")
print(df.groupby('Inflation_Risk')['GDP_Impact_Pct'].mean())


```

```python
df.boxplot(column='GDP_Impact_Pct', by='Inflation_Risk')
plt.title("GDP Impact par Inflation Risk")
plt.show()
```

```python
print("\n--- TABLEAU CROISÉ ---")
print(pd.crosstab(df['Inflation_Risk'], df['Vulnerability']))
```

```python
prob_high = len(df[df['Inflation_Risk']=='High']) / len(df)
print("\nProbabilité Inflation élevée:", prob_high)

prob_cond = len(df[(df['Inflation_Risk']=='High') & (df['Vulnerability']=='Critical')]) / len(df[df['Vulnerability']=='Critical'])
print("Probabilité conditionnelle (High | Critical):", prob_cond)
```

```python
df['Inflation_Risk'].value_counts().plot(kind='bar')
plt.title("Distribution Inflation Risk")
plt.show()
```

```python
print("\n--- INTERPRÉTATION ---")
print("- La majorité des pays ont un risque d’inflation élevé.")
print("- Les pays dépendants du pétrole sont plus impactés.")
print("- Il existe une relation entre vulnérabilité et inflation.")
print("- Les données sont propres et exploitables.")
```
